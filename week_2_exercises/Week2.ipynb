*We import the needed packages*

````python
import pandas as pd
import numpy as np
import csv
````

*We start off by loading the datasets*

````python
data_1 = open("../rawdata/data1.tsv")
data_2 = open("../rawdata/data2.tsv")
data_3 = open("../rawdata/data3.tsv")
data_4 = open("../rawdata/data3.tsv")

data_1 = csv.reader(data_1, delimiter="\t")
data_2 = csv.reader(data_2, delimiter="\t")
data_3 = csv.reader(data_3, delimiter="\t")
data_4 = csv.reader(data_4, delimiter="\t")

all_data = [data_1, data_2, data_3, data_4]


full_list = []
for data in all_data:
    x = []
    y = []
    for row in data:
        x.append(float(row[0]))
        y.append(float(row[1]))
    full_list.append([x,y])
    
data_1, data_2, data_3, data_4 = full_list 

data_1_x, data_1_y = data_1
data_2_x, data_2_y = data_2
data_3_x, data_3_y = data_3
data_4_x, data_4_y = data_4
````

**Using the numpy function mean, calculate the mean of both 洧논-values and 洧녽-values for each dataset.**

````python
data_1_x_mean = np.mean(data_1_x)
data_1_y_mean = np.mean(data_1_y)
data_2_x_mean = np.mean(data_2_x)
data_2_y_mean = np.mean(data_2_y)
data_3_x_mean = np.mean(data_3_x)
data_3_y_mean = np.mean(data_3_y)
data_4_x_mean = np.mean(data_4_x)
data_4_y_mean = np.mean(data_4_y)
````

**The 24 hour cycle. We'll can also forget about weekday and simply count up the number of each crime-type that occurs in the entire dataset from midnight to 1am, 1am - 2am ... and so on**

````python
policing_dataframe['Hour'] = pd.DatetimeIndex(policing_dataframe['Date_Time']).hour
````

*We create a way to mark something in a time-period*

````python
hours_in_day = [i for i in range(0,24)]
hourly_slots = {}
for i in range(len(hours_in_day)):
    if i+1 == len(hours_in_day):
        from_time = hours_in_day[i]
        to_time = hours_in_day[0]
        hourly_slots[from_time] = str(from_time) + "-" + str(to_time)
    else:
        from_time = hours_in_day[i]
        to_time = hours_in_day[i+1]
        hourly_slots[from_time] = str(from_time) + "-" + str(to_time)
    
print(hourly_slots)
````

*We then use a list comprehension that access the dictionary with a key, to then output the time period as a string*

****OK OK, I know Pandas probably has a way to do this, but I had fun creating the above script****

````python
policing_dataframe["time_period"] = [hourly_slots[int(str(i).split(":")[0])] for i in list(policing_dataframe['Date_Time'].dt.hour)]
````

**Next, we want to generate a slightly more complicated graphic. I'm interested to know if there are certain crimes >that happen much more in certain neighborhoods than what's typical. Below I describe how to get that plot going**

*First, we need to calculate the relative probabilities of seeing each type of crime in the dataset as a whole. > That's simply a normalized version of this plot. Let's call it P(crime).*

````python
counts_pr_category = policing_dataframe.groupby(by=["Category"]).size().reset_index(name="counts")
counts_pr_category = counts_pr_category.loc[counts_pr_category['Category'].isin(focuscrimes)].reset_index(drop=True)
counts_pr_category
````

*We create the P(crime) column as a relative probability that is the min-max scalar*

````python
counts_pr_category["P(crime)"] = (counts_pr_category["counts"])/(counts_pr_category["counts"].max())
````

````python
counts_pr_category
````

*Next, we calculate that same probability distribution but for each PD district, let's call that P(crime|district)*

````python
counts_pr_category_pr_district = policing_dataframe.groupby(by=["Category", "PdDistrict"]).size().reset_index(name="counts")
counts_pr_category_pr_district = counts_pr_category_pr_district.loc[counts_pr_category_pr_district['Category'].isin(focuscrimes)].reset_index(drop=True)
````

*We use a pivot table to get the occurences in each district*

````python
pivot_table_types_pr_district = counts_pr_category_pr_district.pivot_table(values='counts', columns='PdDistrict', index=['Category'], aggfunc='sum')
````

````python
districts = ['SOUTHERN', 'MISSION', 'PARK', 'INGLESIDE', 'BAYVIEW', 'RICHMOND',
       'TARAVAL', 'TENDERLOIN', 'NORTHERN', 'CENTRAL']
````

*Like above, we now calculate the relavtive probability based on the district*

````python
for district in districts:
    pivot_table_types_pr_district[district] = (pivot_table_types_pr_district[district])/(pivot_table_types_pr_district[district].max())
````

**Now we look at the ratio P(crime|district)/P(crime). That ratio is equal to 1 if the crime occurs at the same level within a district as in the city as a whole. If it's greater than one, it means that the crime occurs more frequently within that district. If it's smaller than one, it means that the crime is rarer within the district in question than in the city as a whole.**

````python
P_crime_pr_district_city_ratio = pivot_table_types_pr_district.reset_index().copy()
````

````python
import plotly.express as px

fig = px.choropleth_mapbox(df, geojson=data, locations='District', color='Random_data',
                           color_continuous_scale="Viridis",
                           range_color=(0, 1),
                           mapbox_style="carto-positron",
                           zoom=11, center = {"lat": 37.773972, "lon": -122.431297},
                           opacity=0.5,
                           labels={'Random_data':'Random Data'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
````
