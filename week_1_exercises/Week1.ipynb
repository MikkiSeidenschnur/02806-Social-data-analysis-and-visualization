Now it's time to explore how the crime statistics change over time. To start off easily, let's count the number of crimes per year for the years 2003-2017 (since we don't have full data for 2018). What's the average number of crimes per year?

```python
policing_dataframe['Date'] = pd.to_datetime(policing_dataframe['Date'])
```

```python
policing_dataframe['Year'] = pd.DatetimeIndex(policing_dataframe['Date']).year
```

```python
policing_dataframe = policing_dataframe[policing_dataframe.Year != 2018]
```

```python
total_crimes_2003_to_2017 = policing_dataframe['Year'].value_counts().sum()
count_of_years = len(policing_dataframe['Year'].value_counts())

average_crimes_pr_year = int(round(total_crimes_2003_to_2017 / count_of_years, 0))
average_crimes_pr_year
```

Police chief Suneman is interested in the temporal development of only a subset of categories, the so-called focus crimes. Those categories are listed below (for convenient copy-paste action). Now create bar-charts displaying the year-by-year development of each of these categories across the years 2003-2017.

focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])

```python
focuscrimes = ['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT']

counts_pr_category = policing_dataframe.groupby(by=["Category", "Year"]).size().reset_index(name="counts")

counts_pr_category = counts_pr_category.loc[counts_pr_category['Category'].isin(focuscrimes)].reset_index(drop=True)
counts_pr_category
```

```python
plt.figure(figsize=(20,25))
for i in range(len(focuscrimes)):
    temp = counts_pr_category.loc[counts_pr_category['Category'] == focuscrimes[i]].reset_index(drop=True)
    x = temp["Year"]
    y = temp["counts"]
    plt.subplot(len(focuscrimes)/2,2,i+1)
    plt.title(focuscrimes[i])
    plt.bar(x, y)
    plt.tight_layout()
```
