***Difference between data and metadata***

*The main difference is that data describes a phenomenon, like physical measurements of weather data, whilst meta data gives context to data, like describing at which weather station the data was measured.*

***What is the human eye good at***

*Whilst the human eye is good for LOOKING at data, you still need to bring your brain to understand it. The eye can make sense of data that is structured in a way that we can bring our head (brain) around the data.*

***What is the human eye bad at***

*Lifting heavy objects... Just a joke. Our eyes and brain have a tendency to procedurally process data, looking to recognize patterns that makes the previous data make sense - this means that if we do not perceive a pattern, then we cannot comprehend the data. This could for instance be a large feature set for machine learning that has 750 columns - we don't see the connection of this before the data has been structured in a manner that our brains can look for a pattern. Furthermore, our brains hunger to generate and see patterns create a massive bias in data - it gives us a tendency to stop looking for patterns the second we already found one pattern - this pattern might be accidental, but our brain has already released endorphines in our bloodstream, satisfying our OCD to find patterns (OK, OK, that's a long explanation that is more my personal experience than explained science)*

***What is the Simpson's paradox, give an example***

*From my understanding, the simpson's paradox is that the statistical trends for the aggregated data is different from the trends of the subdivided data. This is for instance shown in this [youtube video](https://www.youtube.com/watch?v=t-Ci3FosqZs). When you aggregate data, then you can take for instance 3 seperated datasets that shown and upward trendline seperatedly, but when you aggregate them, the trendline might show as downward, because the 3 data sets are placed differently, but shows the same trend*

***What is the Pearson Correlation, and where does it fail***

*The pearson correlation is a way to look for linear relations in datasets, described by the formula:*

&emsp;
$$r_{pearson} = \displaystyle \frac{n(\sum xy) - \sum x \sum y}{\sqrt{(n\sum x^2 - (\sum x)^2) \times (n\sum y^2) - (\sum y)^2)}    }$$
&emsp;&emsp;

*In general the pearson correlation handles some things poorly: like non-linear relationships - which occurs all the time in the natural world.*
*Furthermore, the pearson correlation has a tendency to be greatly affected by outliers - they can both create a correlation that isn't there, or deduct that there is no correlation because of one point that is marginally different from the sample*

***What is the difference between a bar-chart and a histogram***

*1. Histograms are used to show the distribution of variables whereas bar charts compare several variables*

*2. Histograms bins data while bar charts are categorical*

***How do you chose the right bin-size for a histogram***

*There are a few ways (Quantative & Qualitative) to do it. One method we thought sounds cool is the Sturge's Rule (It has been critized for over smoothing histograms). It's more a rule of thumb than a perfect formula:*

&emsp;
$$K = 1 + 3.322 log_{n}$$
&emsp;&emsp;

*Supposedly it works best for continous data which is normally distributed and symmetrical* ([source](https://www.statisticshowto.com/choose-bin-sizes-statistics/))

***Explain what the point is of a jitter plot***

*The jitter plot assigns a random value on data that is clustered directly on top of each other. This makes it possible to distinguish several data points clustered on top of eachother. It should not be used, for instance for geographical data*

*Plainly, it means that chosing the bin sizes of a histogram can greatly affect what the data looks like. From figure 2 i might assume that the data is distributed evenly, and from figure 3 it's obvious that it's not*

***When are KDE's misleading***

*They may deceive you, since the bandwidth of the KDE is extremely important. If you set the bandwidth too narrow or wide, it may represent your data incorrectly. Also, the KDE plot may seem convincing due to its smooting of the plot*

***What does it mean that "CDFs have less intuitive appeal than histograms of KDEs"?***

*Histograms and KDEs have a tendency to look right from the get-go, whereas cumulative curves can seem less intuitive. This is because they simply take the overall data and sum the total data up. This means that it shows the occurence of datapoints in the total count. I've personally always loved cumulative curves, as they give us a good idea of HOW MUCH of the data is located in a specific part of the dataset*

***What is a quantile plot and what is it good for***

*An example: You have a room indoors. In this room, you want it to be below 26C degrees, for more than 95% of the year (These are the rules for the danish building regulation). This can be evaluated by doing a quantile plot with the temperature measurements over a year, on the y-axis and the percentage on the x-axis. This way we can make sure that this requirement is held, so that people don't fall over from heat exhaustion (Or we spent too much energy on cooling the room) - Therefore I like it for time based data (Or something like spacial data, like square meters)*

***How is a probability plot defined***

*A probability is a graphical way to determine whether or not a data set follows a given distribution. The closer it is to a straight line, the more likely it is to follow that distribution*

***Why should you be careful with using mean, median and general summary statistics***

*As discussed in the previous weekly assignement, statistics can be the same for different datasets, even though the datasets are not the same. For instance the mean value is greatly affected by outliers, which ultimately can result in misrepresenting a dataset, by saying something like: "The mean life expectancy is 84 years". This is misrepresented if that happened because 20 people died at age 180, and 100 people died at age 65*

***When are box plots most useful***

*It is quite useful during explanatory analysis. It can give a good idea on the distribution of the data and whether the median of the data is placed with a positive, negative, or no skewness*

***Are violin plots better/worse than box plots?***

*Violin plots show the multimodal nature of a dataset, whereas box plots do not. That is also why they are called violin-plots - the multimodal..ness of the data set will make it look like a violin*

*We import the necessary packages*

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

*We import the policing dataframe, and parse the date and time as a datetime*

```python
policing_dataframe = pd.read_csv('../files/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv', parse_dates=[['Date', 'Time']])
```

*We assign a mask to select all the data in February 2013*

```python
mask = (policing_dataframe['Date_Time'] > "2013-02-01") & (policing_dataframe['Date_Time'] <= "2013-06-28")
df_assaults = policing_dataframe.loc[mask].reset_index(drop=True)
```

*We specifically select the crimes of category == ASSAULT*

```python
df_assaults = df_assaults.loc[df_assaults['Category'] == "ASSAULT"].reset_index(drop=True)
```

*We filter the data to stuff that happened from 21:00 - 22:00*

*We create a time interval and add it to the df_assaults dataframe*

```python
hours_in_day = [i for i in range(0,24)]
hourly_slots = {}
for i in range(len(hours_in_day)):
    if i+1 == len(hours_in_day):
        from_time = hours_in_day[i]
        to_time = hours_in_day[0]
        hourly_slots[from_time] = str(from_time) + "-" + str(to_time)
    else:
        from_time = hours_in_day[i]
        to_time = hours_in_day[i+1]
        hourly_slots[from_time] = str(from_time) + "-" + str(to_time)
    
print(hourly_slots)
```

```python
df_assaults["time_period"] = [hourly_slots[int(str(i).split(":")[0])] for i in list(df_assaults['Date_Time'].dt.hour)]
```

```python
df_assaults_filtered = df_assaults.loc[df_assaults['time_period'] == "21-22"].reset_index(drop=True)
```

*We create a scatterplot with jitter*

```python
x = df_assaults_filtered["Date_Time"].dt.minute
```

```python
plt.figure(figsize=(10,10))
ax = sns.stripplot(x=x, jitter=True)
```

***We now want to create a boxplot that shows the crimes distributed over the day***

*We start of by creating a function that can calculate and add the time since midnight*

```python
def seconds_since_midnight(dataframe):
    from_time = pd.Timestamp("00:00:00")
    to_time = dataframe.time()
    return to_time.hour * 3600 + to_time.minute * 60
```

*We apply above function*

```python
policing_dataframe["Time since midnight"] = policing_dataframe.Date_Time.apply(seconds_since_midnight)
```

*We filter by the time of day and category*

```python
time_of_day_pr_category = policing_dataframe.groupby(by=["Category", "Time since midnight"]).size().reset_index(name="counts")
```

*We create the datasets seperately*

```python
data = []
for crime in focuscrimes:
    temp = time_of_day_pr_category.loc[time_of_day_pr_category['Category'] == crime].reset_index(drop=True)
    data.append(temp['Time since midnight'])
```

*We create the y-axis labels*

```python
seconds_pr_hour = [3600*i for i in range(0,25)]
```

*We now plot the data in a box-plot*
